# mart_city_top_products

Витрина топ-2 товаров по выручке по городам

```scala
%spark.pyspark
from pyspark.sql import functions as F
from pyspark.sql.window import Window

users = spark.createDataFrame(
    [
        ("u1", "Berlin"),
        ("u2", "Berlin"),
        ("u3", "Munich"),
        ("u4", "Hamburg"),
    ],
    ["user_id", "city"],
)

orders = spark.createDataFrame(
    [
        ("o1", "u1", "p1",  2, 10.0),
        ("o2", "u1", "p2",  1, 30.0),
        ("o3", "u2", "p1",  1, 10.0),
        ("o4", "u2", "p3",  5,  7.0),
        ("o5", "u3", "p2",  3, 30.0),
        ("o6", "u3", "p3",  1,  7.0),
        ("o7", "u4", "p1", 10, 10.0),
    ],
    ["order_id", "user_id", "product_id", "qty", "price"],
)

products = spark.createDataFrame(
    [
        ("p1", "Ring VOLA"),
        ("p2", "Ring POROG"),
        ("p3", "Ring TISHINA"),
    ],
    ["product_id", "product_name"],
)

users.show()
orders.show()
products.show()
```

### 1. revenue = qty * price

```scala
%spark.pyspark
orders_rev = orders.withColumn("revenue", F.col("qty") * F.col("price"))
orders_rev.show()
```

### 2. join с users и products

```scala
%spark.pyspark
enriched = (
    orders_rev
    .join(users, "user_id")
    .join(products, "product_id")
)
enriched.show()
```

### 3. агрегаты по (city, product_id, product_name)

```scala
%spark.pyspark
agg = (
    enriched
    .groupBy("city", "product_id", "product_name")
    .agg(
        F.count("order_id").alias("orders_cnt"),
        F.sum("qty").alias("qty_sum"),
        F.sum("revenue").alias("revenue_sum"),
    )
)
agg.orderBy("city", F.desc("revenue_sum")).show()
```

### 4. top-2 товара по городу


```scala
%spark.pyspark
w = Window.partitionBy("city").orderBy(F.desc("revenue_sum"))

mart = (
    agg
    .withColumn("rn", F.row_number().over(w))
    .filter(F.col("rn") <= 2)
    .drop("rn")
)

mart.orderBy("city", F.desc("revenue_sum")).show()
```

### 5. запись в HDFS и S3


```scala
%spark.pyspark
hdfs_path = "/tmp/sandbox_zeppelin/mart_city_top_products/"
s3_path   = "s3a://seminar-storage/mart_city_top_products/"

mart.write.mode("overwrite").parquet(hdfs_path)
mart.write.mode("overwrite").parquet(s3_path)
```

### 6. чтение обратно


```scala
%spark.pyspark
spark.read.parquet(hdfs_path).orderBy("city", F.desc("revenue_sum")).show()
spark.read.parquet(s3_path).orderBy("city", F.desc("revenue_sum")).show()
```
